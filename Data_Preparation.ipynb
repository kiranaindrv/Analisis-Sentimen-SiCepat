# Data Preparation

## Pre-processing

### Data Cleaning

```python
df = filtered_data[['full_text']]
df
```

```python
# menghindari tweet duplikat/spam
df = df.drop_duplicates()
df
```

#### regex

```python
import regex as re

def clean_twitter_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text) #penghapusan tag mention/user
    text = re.sub(r'#\w+', '', text) #penghapusan hashtag
    text = re.sub(r'RT[\s]+', '', text) #menghapus retweet/RT
    text = re.sub(r'https://\S+', '', text) #penghapusan url
    text = re.sub(r'[^A-Za-z0-9 ]+', '', text) #penghapusan karakter non alphanumeric
    text = re.sub(r'\s+', ' ', text).strip() #penghapusan spasi berlebih
    text = re.sub(r"\d+", "", text) # penghapusan angka

    return text
```

```python
df['full_text'] =  df['full_text'].apply(clean_twitter_text)
df
```

#### Abbreaviation Removal

```python
# Membaca kamus normalisasi singkatan dari file CSV
kamus_df = pd.read_csv('normalisasi.csv', header=None, names=['singkatan', 'lengkap'])

# Mengubah DataFrame kamus menjadi dictionary
kamus_singkatan = pd.Series(kamus_df.lengkap.values, index=kamus_df.singkatan).to_dict()

# Fungsi untuk meng-expand singkatan berdasarkan kamus normalisasi
def expand_singkatan(text, kamus):
    words = text.split()
    expanded_words = [kamus[word] if word in kamus else word for word in words]
    return ' '.join(expanded_words)
```

```python
df['full_text'] = df['full_text'].apply(lambda x: expand_singkatan(x, kamus_singkatan))
df.head(15)
```

#### transalte indo -> inggris

```python
# Fungsi untuk menghapus kata tertentu dari teks
def remove_word(text, word):
    return ' '.join([w for w in text.split() if w.lower() != word.lower()])
```

```python

# Menerapkan fungsi untuk menghapus kata pada kolom 'full_text', penghilangan brand ekspedisi dengan tujuan menghilangkan bias saat teks ditranslate
df['cleaned_text'] = df['full_text'].apply(lambda x: remove_word(x, 'sicepat'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'anteraja'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'jne'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'jnt'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'wahana'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'shopee express'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'ninja express'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'gosend'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'pos'))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_word(x, 'spx'))

# Menampilkan DataFrame
df
```

```python
from googletrans import Translator

# Inisialisasi penerjemah
translator = Translator()

# Fungsi untuk menerjemahkan teks dari bahasa Indonesia ke bahasa Inggris
def translate_text(text, src='id', dest='en'):
    try:
        translated = translator.translate(text, src=src, dest=dest)
        return translated.text
    except Exception as e:
        return str(e)
```

```python
# Menerapkan fungsi translate pada kolom 'full_text'
df['translated_text'] = df['cleaned_text'].apply(translate_text)

# Menampilkan DataFrame
df
```
