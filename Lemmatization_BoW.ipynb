### Punctuation Repetition Removal

```python
import regex as re

def punctuation_repetition(text, default_replace=""):
  text= re.sub(r'[\?\.\!]+(?=\?\.\!)', default_replace, text)
  return text
```

```python
def word_repetition(text):
  text = re.sub(r'(.)\1+', r'\1\1', text)
  return text
```

### Stopword Removal

```python
%pip install NLTK
```

```python
# Stopword Removal atau menghapus kata yang tidak memiliki arti

import nltk
from nltk.corpus import stopwords

# Unduh stopwords
nltk.download('stopwords')

# Ambil daftar stopwords untuk bahasa Inggris
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
pd.set_option('display.max_colwidth', None)
twt_translate.head(20)
```

```python
twt_translate['stopwords'] = twt_translate['translated_text'].apply(lambda x: remove_stopwords(x))
twt_translate
```

```python
import regex as re

def clean_twitter_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text) #penghapusan tag mention/user
    text = re.sub(r'#\w+', '', text) #penghapusan hashtag
    text = re.sub(r'RT[\s]+', '', text) #menghapus retweet/RT
    text = re.sub(r'https://\S+', '', text) #penghapusan url
    text = re.sub(r'[^A-Za-z0-9 ]+', '', text) #penghapusan karakter non alphanumeric
    text = re.sub(r'\s+', ' ', text).strip() #penghapusan spasi berlebih
    text = re.sub(r"\d+", "", text) # penghapusan angka

    return text
```

```python
twt_translate['stopwords'] =  twt_translate['stopwords'].apply(clean_twitter_text)
twt_translate
```

### Normalization: Lemmatization

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Unduh dataset yang diperlukan
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Inisialisasi lemmatizer
lemmatizer = WordNetLemmatizer()

# Fungsi untuk mendapatkan POS tag yang sesuai untuk lemmatizer
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Fungsi untuk melakukan lemmatization pada token-token
def lemmatize_tokens(tokens):
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]
    return lemmatized_tokens
```

```python
twt_translate['tokens_lemmatized'] = twt_translate['tweet_tokenize'].apply(lemmatize_tokens)
twt_translate.tail(10)
```

## Data Labeling

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Kamus kecil manual (kata dan skor sentimen kustom)
custom_words = {
    'slow': -0.5,      
    'fast': 0.5,  
}

# Inisialisasi analyzer
analyzer = SentimentIntensityAnalyzer()

# Menambahkan kamus kecil manual ke analyzer
analyzer.lexicon.update(custom_words)

# Fungsi untuk mendapatkan skor sentimen tanpa nilai neutral dan menambahkan label
def get_sentiment_with_label(text):
    sentiment_dict = analyzer.polarity_scores(text)
    
    # Hilangkan nilai neutral
    pos = sentiment_dict['pos']
    neg = sentiment_dict['neg']
    neu = sentiment_dict['neu']
    
    # Tentukan label berdasarkan nilai compound
    if sentiment_dict['compound'] >= 0.1:
        label = 'positive'
    elif sentiment_dict['compound'] < 0:
        label = 'negative'
    else:
        label = 'neutral'
    
    return {
        'pos': pos,
        'neg': neg,
        'neu': neu,
        'label': label
    }
```

```python
# Fungsi untuk menggabungkan token menjadi satu kalimat
def join_tokens(tokens):
    return ' '.join(tokens)

# Gabungkan token dalam setiap baris menjadi kalimat
df['text_lemmatized'] = df['tokens_lemmatized'].apply(join_tokens)
```

# Modelling dan Evaluasi

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Encode label teks menjadi numerik
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['label'])
```

```python
print(X_vec)
```

### Model (data split A) train:test -> 90:10 [Hasil Akurasi: 0, 863]

```python
# Membagi data menjadi set pelatihan dan pengujian (90:10)
X_train_A, X_test_A, y_train_A, y_test_A = train_test_split(X_vec, df['label_encoded'], test_size=0.1, random_state=42)
```

```python
# Definisikan model SVM
svm = SVC()

# Tentukan grid hyperparameter
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search.fit(X_train_A, y_train_A)

# Menampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

```python
# Prediksi dengan model terlatih
y_pred_A = grid_search.predict(X_test_A)

# Menghitung metrik evaluasi
accuracy = accuracy_score(y_test_A, y_pred_A)
precision = precision_score(y_test_A, y_pred_A, average='weighted')
recall = recall_score(y_test_A, y_pred_A, average='weighted')
f1 = f1_score(y_test_A, y_pred_A, average='weighted')

print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-Score: {f1:.3f}")

# Buat confusion matrix
cm = confusion_matrix(y_test_A, y_pred_A)

# Konversi confusion matrix ke DataFrame untuk visualisasi yang lebih baik
cm_df = pd.DataFrame(cm, index=label_encoder.classes_, columns=label_encoder.classes_)

# Cetak laporan klasifikasi
report = classification_report(y_test_A, y_pred_A, target_names=label_encoder.classes_)
print("\nClassification Report:\n", report)

# Tampilkan confusion matrix sebagai grafik
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```
